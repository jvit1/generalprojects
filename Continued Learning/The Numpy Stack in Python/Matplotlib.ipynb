# Line Chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1000)
```

```python
y = np.sin(x) + .2 * x
```

```python
plt.plot(x,y)
```

```python
# Label x axis
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('myplot');  # Hide output
```

# Scatterplot

```python
# Let's create some random data
X = np.random.randn(100,2)
plt.scatter(X[:,0], X[:,1]);
```

```python
X = np.random.randn(200,2)
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:,0], X[:,1], c=Y);
```

# Exercise
We need to generate and plot the XOR dataset.

```python
x1 = (np.random.random(1000) - 0.5) 
x2 = (np.random.random(1000) - 0.5) 
Y= np.zeros(1000)

for i in range(len(Y)):
    if (x1[i] <= 0 and x2[i] > 0) or (x1[i] > 0 and x2[i] <= 0):
        Y[i] = 1
```

```python
plt.scatter(x1,x2,c=Y);
```
